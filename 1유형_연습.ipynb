### Q19. 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
result = video.sort_values(['videoname', 'ct'])[video.ct >= pd.to_datetime('2021-11-01 00:00:00')].drop_duplicates('videoname')
result.reset_index(drop=True, inplace=True)
```

```python
result['viewcnt2'] = video.sort_values(['videoname', 'ct'])[video.ct <= pd.to_datetime('2021-11-01 15:00:00')].drop_duplicates('videoname', keep='last')['viewcnt'].reset_index(drop=True)
result[['viewcnt', 'viewcnt2']]
```

```python
result['viewcnt_증가량'] = result['viewcnt2'] - result['viewcnt']
display(result[['videoname', 'viewcnt_증가량']])
```

```python
## 답
start = pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

target = video.loc[(video["ct"] >= start) & (video['ct'] <= end)].reset_index(drop=True)

def check(x):
    result = max(x) - min(x)
    return result

answer = target[['videoname','viewcnt']].groupby("videoname").agg(check)
answer
```
